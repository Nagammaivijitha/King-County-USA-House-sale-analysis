## Create a correlation heatmap using seaborn

```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(15,15))

# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.
corr = sns.heatmap(df_kc_house_corrl_data.corr(), annot = True, ax = ax) # The `annot` argument allows the plot to 
#place the correlation coefficients onto the heatmap.
plt.savefig("seaborn Correlation Matrix.png") 
```

## Discuss what the coefficients in the plot mean in terms of the relationships between the variables.

```python
# Here i mainly focus on price correlation with other variables.
# 3 variables have strong positive correlation with price.
```

```python
# Price and Sqft_living: There is a strong positive correlation of 0.702 between the price of a house and its square footage
# of living space (Sqft_living). This indicates that larger houses tend to have higher prices.
```

```python
# Price and Grade: There is a strong positive correlation of 0.667 between the price of a house and its grade. 
# Grade typically represents the overall quality of construction and design of the house. 
# Therefore, houses with higher grades tend to have higher prices.
```

```python
# Price and Sqft_Above: There is a strong positive correlation of 0.606 between the price of a house and the square footage above ground level (Sqft_Above). 
# This suggests that larger above-ground living areas contribute to higher house prices.
```

```python
# Price and Bathrooms: There is a moderate positive correlation of 0.510 between the price of a house and the number of bathrooms it has.
# This suggests that houses with more bathrooms tend to have higher prices.
```

```python
# Price and View: There is a moderate positive correlation of 0.397 between the price of a house and its view. 
# This implies that houses with better views tend to have higher prices.
```

# Sactterplots

```python
# Creating scatter plots with regression lines
sns.set(style="ticks")
sns.lmplot(x='Sqft_living', y='Price', data=df_kc_house_corrl_data)
plt.title('Price vs Sqft_living')
plt.savefig("scatter plot Price vs Sqft_living.png") 

sns.lmplot(x='Grade', y='Price', data=df_kc_house_corrl_data)
plt.title('Price vs Grade')
plt.savefig("scatter plot price vs Grade.png") 

sns.lmplot(x='Sqft_Above', y='Price', data=df_kc_house_corrl_data)
plt.title('Price vs Sqft_Above')
plt.savefig("scatter plot price vs Sqft_Above.png") 
plt.show()
```

```python
# the scatterplot clears that the price value is high and depends on sqft_living, sqft_above, grade. 
```

```python
df_kc_house_corrl_data.loc[df_kc_house_corrl_data['Price'] < 321950.00000, 'Price category'] = 'Low price'
df_kc_house_corrl_data.loc[(df_kc_house_corrl_data['Price'] >= 321950.00000) & (df_kc_house_corrl_data['Price'] < 645000.00000), 'Price category'] = 'Middle price'
df_kc_house_corrl_data.loc[df_kc_house_corrl_data['Price'] >= 645000.00000, 'Price category'] = 'High price'
```
