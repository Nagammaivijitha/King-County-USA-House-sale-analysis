# # Changing the Variable’s Data Type

```python
# Here i am changing 5 variables datatype. Date from object to datetime format, Bathrooms from float to int64, Floors from float to int64, Year_Build from int to datetime , Year_Renovated from int to datetime.
```

```python
# changing Date variable datatype from object to datetime.
df_kc_house_data2['Date'] = pd.to_datetime(df_kc_house_data2['Date'])
```

```python
# changing Bathrooms variable datatype from float to int64.
df_kc_house_data2['Bathrooms'] = df_kc_house_data2['Bathrooms'].astype('int64')
```

```python
# changing Floors variable datatype from float to int64.
df_kc_house_data2['Floors'] = df_kc_house_data2['Floors'].astype('int64')
```

```python
# changing Year_Built variable datatype from int to datetime.
df_kc_house_data2['Year_Built'] = pd.to_datetime(df_kc_house_data2['Year_Built'])
```

```python
# changing Year_Renovated variable datatype from int to datetime.
df_kc_house_data2['Year_Renovated'] = pd.to_datetime(df_kc_house_data2['Year_Renovated'])
```

```python
# checking whether the variables are changed to correct datatype.
df_kc_house_data2.dtypes
```
